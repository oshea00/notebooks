```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# if not jupyter - use plt.show() to display charts

```python
x = np.linspace(0,5,11)
y = x ** 2
```

# Functional method of plotting

```python
plt.plot(x,y,'r--'); # matlib-like parameters
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Example Chart')
```

# Multiple plots

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b');

```

# Object-oriented style

```python
fig = plt.figure() # obtain a figure object
axes = fig.add_axes([0,0,1,1]) # [left,bottom,width,height]
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Example Chart');
```

## Multiple plots (with their own axes) on same figure

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.set_title('Example With Embedded Chart')
axes2 = fig.add_axes([.1,.6,.25,.25])
axes2.set_title('Zoom')
```

## Multiple sub-plots on figure

```python
fig, axes = plt.subplots(2,figsize=(5,5))
fig.set_figwidth(10) # overrides figsize
fig.set_figheight(6) 
axes[0].plot(x,y)
axes[0].set_title('Subplot 1')
axes[1].plot(y,x)
axes[1].set_title('Subplot 2')
plt.tight_layout(); # updates layout so sub-plots don't overlap
```

## rows and columns - nifty style options

```python
#plt.style.use('seaborn')
with plt.xkcd():
    fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
    axes[0][0].plot(x,y)
    axes[0][0].set_title('Subplot 1',color='black')
    axes[0][0].annotate('*Source: math',xy=(3.8,1.5),xytext=(3.8,1.5)) # help(plt.annotate) for more options
    axes[0][1].plot(x,x**2,label='y=x^2')
    axes[0][1].plot(x,x**3,label='y=x^3')
    axes[0][1].legend(loc='upper left', shadow=True, fontsize='small')
    axes[0][1].set_title('Subplot 2 - multiple data series',color='black')
    axes[1][0].pie(y)
    axes[1][0].set_title('Subplot 3',color='black')
    axes[1][1].bar(x,y,width=0.3)
    axes[1][1].set_title('Subplot 4',color='black')
    my_suptitle = plt.suptitle('All My Charts',y=1.1,color='black')
    plt.tight_layout(); # updates layout so sub-plots don't overlap
    # Save output
    fig.savefig('mychart.png', dpi=90, bbox_inches='tight',bbox_extra_artists=[my_suptitle])
```

# Matrix Plots - Magic Stuff!!!

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
flights.head()
```

```python
tips.corr() # a 'matrix' form of data
```

```python
tc = tips.corr()
sns.heatmap(tc,annot=True,cmap='coolwarm');
```

```python
fp =flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
plt.figure(figsize=(10,8))
plt.title('Historical Passenger Volume',y=1.02)
sns.heatmap(fp,cmap='coolwarm',linecolor='white',linewidth=1); # nifty line color!
```

```python
dayvtime = tips.pivot_table(index='day',columns='time',values='tip')
```

```python
sns.heatmap(dayvtime,annot=True,cmap='magma');
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1); # attempt to cluster based on similar rows and columns
```

# Customize using PairGrid

```python
g = sns.PairGrid(iris)  # OMG
#g.map(plt.scatter);
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
