```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer.DESCR)
```

```python
df = pd.DataFrame(data=cancer['data'],columns=cancer['feature_names'])
```

```python
sns.heatmap(df.corr())
```

```python
df.head()
```

```python
cancer.target_names
```

```python
# Scale data 
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
help(StandardScaler)
```

```python
scaler = StandardScaler()
print(scaler.fit(df))
```

```python
np.shape(scaler.transform(df))
```

```python
dfs = pd.DataFrame(data=scaler.transform(df),columns=df.columns)
```

```python
dfs.head()
```

```python
from sklearn.decomposition import PCA
```

```python
help(PCA)
```

```python
pca = PCA(n_components=2)
pca.fit(dfs)
```

```python
pca.singular_values_
```

```python
dft = pd.DataFrame(pca.transform(dfs))
```

```python
sns.pairplot(data=dft)
```

```python
dft['target'] = pd.Series(cancer.target)
```

```python
dft.columns = ['X','Y','target']
```

```python
sns.lmplot( x='Y', y="X", data=dft, fit_reg=False, hue='target')
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_,columns=cancer['feature_names'])
```

```python
df_comp.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma')
```
