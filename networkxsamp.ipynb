```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def greedy_coloring(graph, num_colors=4):
  """
  This function uses a greedy algorithm to color a graph with the given number of colors, ensuring that no two neighboring nodes have the same color.

  Parameters:
    graph: A networkx graph.
    num_colors: The number of colors to use.

  Returns:
    A dictionary representing the coloring of the graph. The keys are the nodes and the values
    are the colors.
  """

  # Initialize the coloring.
  coloring = {}
  for node in graph.nodes():
    coloring[node] = None

  # Iterate over the nodes in the graph.
  for node in graph.nodes():
    # If the node has not been colored yet, color it with the first available color.
    if coloring[node] is None:
      available_colors = set(range(num_colors))
      for adjacent_node in graph.neighbors(node):
        if coloring[adjacent_node] is not None:
          available_colors.remove(coloring[adjacent_node])
      coloring[node] = min(available_colors)

  # Return the coloring.
  return coloring


# Create a graph.
graph = nx.Graph()
graph.add_edges_from([("A", "B"), ("A", "C"), ("B", "C"), ("B", "D"), ("C", "D"), ("C", "E"), ("D", "E"), ("E","G"),("G","C"),("G","A")])

# Color the graph.
coloring = greedy_coloring(graph, num_colors=4)

# Plot the graph.
nx.draw(graph, node_color=[coloring[node] for node in graph.nodes()], with_labels=True)
plt.show()
```
