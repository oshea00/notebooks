```python
datesidx = pd.date_range(start='2017-01-01',end='2019-01-20')
```

```python
prices = []
for dt in [d.strftime('%Y-%m-%d') for d in datesidx]:
    prices.append(client.get_spot_price(currency_pair='BTC-USD',date=dt))
    print(dt)
```

```python
aprices = np.array([float(p['amount']) for p in prices])
dfprices = pd.DataFrame(data=aprices,index=datesidx,columns=['BTC-USD'])
```

```python
dfprices.plot(kind='hist',bins=50);
plt.title('Prices')
plt.legend();
```
