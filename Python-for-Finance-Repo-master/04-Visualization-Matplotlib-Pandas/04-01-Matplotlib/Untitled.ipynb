```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.arange(1,20) 
```

```python
y = x**2
```

```python
plt.plot(x,y,'dr')
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("My Plot")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([.1,.1,.8,.8])
axes.plot(x,y)
axes.set_xlabel("X Vals")
axes.set_ylabel("Y Vals")
axes.set_title("My Plot")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([.1,.1,.8,.8])
axes2 = fig.add_axes([.2,.5,.3,.3])
axes1.plot(x,y)
axes1.set_title("My Data")
axes1.set_xlabel("Days of Study")
axes1.set_ylabel("Knowledge")
axes2.plot(y,x)
axes2.set_title("Zoom")
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x,'r')
axes[1].set_ylabel("Days of Study")
fig.tight_layout()
axes = fig.add_axes([.6,.2,.2,.2])
```

```python
fig = plt.figure(figsize=(8,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
fig.tight_layout()
```

```python
fig.savefig('mypic.png')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,label='x Squared')
axes.plot(x,x**3,label='y Cubed')
axes.legend(loc=(.3,.3))
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlim([1,5])
axes.set_ylim([0,20])
```
